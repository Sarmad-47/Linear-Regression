#### Importing the libraries

```python
import numpy as np 
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

### Loading the data

```python
raw_data=pd.read_csv('Car Sales.csv')
raw_data.head()
```

### Preprocessing

#### Exporing the descriptive statistics of the variables

```python
raw_data.describe(include='all')
```

### Determining the variables of interest

```python
# we shall drop the Model column because most values are null. So we must clean the data first.
# axis=0 means rows
# axis = 1 means columns
data=raw_data.drop(['Model'],axis=1)
data.describe(include='all')
```

#### Dealing with missing values

```python
data.isnull().sum()
```

```python
# Dropping those values with ad hoc method
data_no_mv = data.dropna(axis=0)
```

```python
data_no_mv.describe(include='all')
```

### Exploring the PDFs

```python
sns.distplot(data_no_mv['Price'])
```

#### Dealing with outliers

**Dealing with price**

```python
q=data_no_mv['Price'].quantile(0.99)
data_1=data_no_mv[data_no_mv['Price']<q]
data_1.describe(include='all')
```

```python
sns.distplot(data_1['Price'])
```

**Dealing with mileage**

```python
sns.distplot(data_no_mv['Mileage'])
```

```python
q=data_1['Mileage'].quantile(0.99)
data_2=data_1[data_1['Mileage']<q]
sns.distplot(data_2['Mileage'])
```

**Dealing with Engine Volume**

```python
sns.distplot(data_no_mv['EngineV'])
```

```python
data_3=data_2[data_2['EngineV']<6.5]
sns.distplot(data_3['EngineV'])
```

**Dealing with the year**

```python
sns.distplot(data_no_mv['Year'])
```

```python
q=data_3['Year'].quantile(0.01)
data_4=data_3[data_3['Year']>q]
sns.distplot(data_4['Year'])
```

```python
data_cleaned=data_4.reset_index(drop=True)
```

```python
data_cleaned.describe(include='all')
```

## Checking the OLS assumptions

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')


plt.show()
```

```python
sns.distplot(data_cleaned['Price'])
```

#### Relaxing the assumptions

```python
## We are doing this to make a linear pattern
log_price = np.log(data_cleaned['Price'])
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['log_price'])
ax3.set_title('Log Price and Mileage')


plt.show()
```
